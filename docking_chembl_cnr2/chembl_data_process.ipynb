```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("cnr2_ki.csv", delimiter=";")
```

```python
data.head()
```

```python
data_cleaned = data[["Molecule ChEMBL ID","Smiles","Standard Value"]]
data_cleaned.info()
```

```python
data_cleaned.dropna(inplace=True)
```

```python
data_cleaned.info()
```

```python
data_cleaned.rename(columns = {"Molecule ChEMBL ID" : "ID", "Standard Value" : "Ki"}, inplace=True)
data_cleaned.head()
```

```python
data_cleaned.describe()
```

```python
data_cleaned.drop(data_cleaned[data_cleaned['Ki'] > 1200].index, inplace=True)
```

```python
data_cleaned.describe()
```

```python
binwidth = 100
data_cleaned.plot.hist(bins=range(np.min(data_cleaned['Ki'].astype(int)), np.max(data_cleaned['Ki'].astype(int)+ binwidth), binwidth))
```

```python
with open("inhibitors_smiles_list.smi", "a") as f_out:
    for idx, row in data_cleaned.iterrows():
        smi = row["Smiles"]
        id = row["ID"]
        f_out.write(f"{smi}\t{id}\n")
```
